```python
from pydantic import BaseModel
from typing import List

class classroom(BaseModel):
    RoomNumber:int
    students:List[str]
    capacity:int

```

```python
class_boject1=classroom(RoomNumber=1,students=['Parth','Tom','Hansen'], capacity=50)
```

```python
try:
    classroom(RoomNumber=1, students=['Parth', 23, 'Hansen'], capacity=50)
except ValueError as e:
    print(e)
```

Model with Nested Models

Creating complex structure with models

```python
from pydantic import BaseModel

class Address(BaseModel):
    street:str
    city:str
    zip: int

class Customer(BaseModel):
    customer_id:int
    customer_name:str
    address:Address
```

```python
customer1=Customer(
    customer_id=1,
    customer_name='Parth',
    address={"street":"lp ln","city":"r hill","zip":"06067"}
)
```
