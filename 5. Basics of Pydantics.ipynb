Pydantic is a Python library for data validation and settings management that uses type hints to define the structure and types of data. It automatically parses and validates incoming data against these types, converting them when possible and raising clear errors when the data is invalid.

```python
from pydantic import BaseModel
```

```python
class person(BaseModel):
    Name:str
    Age:int
    City:str
```

```python
person_object1=person(Name="Parth",Age=20,City="Indore")
```

```python
print(person_object1)
```

```python
type(person_object1)
```

Now Let's see if we assign values which can not be typecasted

```python
person_object2=person(Name=12, Age=20, City='Pune')
```

Lets understand how pydantic is different than dataclass

Here Data class attribute Name, city assigned intiger value even though they were not defined as str. It did not give error. Hence Pydantic is preferred. 

```python
from dataclasses import dataclass

@dataclass
class person_dataclass():
    Name:str
    Age:int
    City:str
    
```

```python
person_dataclass_object1=person_dataclass(Name=23, Age=34, City=50)
```
