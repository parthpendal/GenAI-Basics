Pydantic with optional fields

```python
from pydantic import BaseModel
```

```python
from typing import Optional
```

```python
class Employee(BaseModel):
    Emp_id:int
    Emp_name:str
    Emp_dept:Optional[str]=None
    Is_active:Optional[bool]=True
```

```python
emp1=Employee(Emp_id=1,Emp_name='Parth')
emp2=Employee(Emp_id=1,Emp_name='Parth',Emp_dept='CIS',Is_active=False)
```

```python
emp1
```

```python
emp2
```

Definition:

Optional[type]: Indicates the field can be None

We can even have list of strings

```python
from pydantic import BaseModel
from typing import List

class classroom(BaseModel):
    RoomNumber:int
    students:List[str]
    capacity:int

```

```python
class_boject1=classroom(RoomNumber=1,students=['Parth','Tom','Hansen'], capacity=50)
```

```python
class_boject1
```

```python
try:
    classroom(RoomNumber=1, students=['Parth', 23, 'Hansen'], capacity=50)
except ValueError as e:
    print(e)
```
